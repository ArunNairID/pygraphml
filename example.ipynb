```python
%matplotlib inline

import tempfile
import os
import sys

from pygraphml import GraphMLParser
from pygraphml import Graph

tmp_dir = tempfile.gettempdir()
```

```python
# Create graph

g = Graph()

n1 = g.add_node("A")
n2 = g.add_node("B")
n3 = g.add_node("C")
n4 = g.add_node("D")
n5 = g.add_node("E")

g.add_edge(n1, n3)
g.add_edge(n2, n3)
g.add_edge(n3, n4)
g.add_edge(n3, n5)

parser = GraphMLParser()
graph_file = os.path.join(tmp_dir, "test.graphml")
parser.write(g, graph_file)

g.show(show_label=True)
```

```python
# Read graph
parser = GraphMLParser()
g = parser.parse(graph_file)

g.show()
```
